```python
from collections import defaultdict
```

```python
def formater(string):
    case, event = string.replace("\n","").split(",")
    return case, event
def process_logs(arch):
    logs = defaultdict(list)
    events = set()
    with open(arch, "r") as arch:
        for i in arch.readlines()[1:]:
            case, event = formater(i)
            logs[case].append(event)
            events.add(event)
    return events, logs
        
```

```python
def zero_matrix_events(events):
    ev = list(events)
    ev.sort()
    event_index = {}
    cont = 0
    for i in ev:
        event_index[i] = cont
        cont += 1
    matrix = []
    for i in range(len(ev)):
        matrix.append([0]*len(ev))
    return event_index, matrix
```

```python
events, logs = process_logs("TDlog.csv")
def alpha_miner_window(events, logs, window):
    event_index, matrix = zero_matrix_events(events)
    for log in logs.values():
        for index in range(0,len(log)-window):
            value1 = log[index]
            value2 = log[index+window]
            index1 = event_index[value1]
            index2 = event_index[value2]
            matrix[index1][index2] += 1
    return matrix
window = 1
alpha_miner_window(events, logs, window)
```
