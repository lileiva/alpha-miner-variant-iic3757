```python
from collections import defaultdict
```

```python
def formater(string):
    case, event = string[:-1].split(",")
    return case, event
logs = defaultdict(list)
events = set()
with open("TDlog.csv", "r") as arch:
    for i in arch.readlines()[1:]:
        case, event = formater(i)
        logs[case].append(event)
        events.add(event)
        
```
